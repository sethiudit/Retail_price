# T-testing

```python
filtered_values = df[(df.State=="Punjab") & (df.Commodity=="Onion") & (df.Unit=="Kg.") & (df.Date=="Jan-2001") & df['Retail Price']]
filtered_values
```

```python
filtered_values['Retail Price'].mean()
```

```python
#filtered_values = display(df.loc[(df['State'].str.startswith('Rajasthan')) & (df['Commodity'].str.startswith('Rice')) & (df['Unit'].str.startswith('Kg.')) & (df['Date'].str.startswith('Jan-2001')) & df['Retail Price'],['State','Commodity','Unit','Date','Retail Price']])
#filtered_values

tscore, pscore = st.ttest_1samp(filtered_values['Retail Price'], popmean=10)
one_tailed_p_value=float("{:.6f}".format(pscore/2)) 
print('Test statistic is %f'%float("{:.6f}".format(tscore)))
print('p-value for one_tailed_test is %f'%one_tailed_p_value)
if one_tailed_p_value<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```

```python
df2 = df['Retail Price'].fillna(df['Retail Price'].median(),inplace = True)
filtered_values1 = df[(df.Commodity=="Milk") & (df.Unit=="Litre") & (df.Date=="Nov-2019") & df['Retail Price']]
filtered_values1
```

```python
filtered_values1['Retail Price'].mean()
```

```python
from statsmodels.stats import weightstats as stests
ztest ,propability_value = stests.ztest(filtered_values1['Retail Price'], x2=None, value=40)
print(float(propability_value))
if propability_value<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```
